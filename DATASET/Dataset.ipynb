```python
import os
import shutil
import cv2
import numpy as np  
from scipy.ndimage import rotate
import matplotlib.pyplot as plt
```

```python
def choose_random_images(path, quantity, move_to = -1):
    images_path = np.array(os.listdir(path))
    my_data_set = np.random.choice(images_path, quantity)
    for image_path in my_data_set:
        newPath = shutil.copy(path + image_path, './image/')
    return my_data_set
```
