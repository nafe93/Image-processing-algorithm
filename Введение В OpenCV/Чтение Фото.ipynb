```python
import cv2
import matplotlib.pyplot as plt
```

### Чтение Картинки

```python
read_img = cv2.imread("../DATA/3.jpg")
```

```python
def view_image(img):
    convert_image_to_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(convert_image_to_RGB)
```

```python
view_image(read_img)
```
