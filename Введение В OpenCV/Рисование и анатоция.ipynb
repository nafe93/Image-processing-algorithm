```python
import cv2
import matplotlib.pyplot as plt
```

```python
def view_image(img):
    convert_image_to_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(convert_image_to_RGB)
```

```python
img = cv2.imread('../DATA/1.jpg')
```

```python
view_image(img)
```

```python
width = int(img.shape[0] / 2)
heigh = int(img.shape[1] / 2)
```

```python
img = cv2.circle(img, center=(width, heigh) , radius=100, color=(0, 0 ,255), thickness=1, lineType=50, shift=0)
```

```python
view_image(img)
```
