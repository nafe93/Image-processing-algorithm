```python
from scipy import ndimage
import numpy as np
import cv2

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def discrete_radon_transform(img, steps = 180):
    # crate matrix
    w, h = img.shape
    zero = np.zeros((h, steps), dtype='float64')
    # roatate
    for s in range(steps):
        rotation = ndimage.rotate(img, s, reshape=False).astype('float')
    # sum
        zero[:, s] = np.sum(rotation, axis=0)
    # rotate image
    zero = ndimage.rotate(zero, 180, reshape=False).astype('float')
    return zero
```

```python
def calculate_rotate_h(img):
    r = np.array([np.sqrt(np.mean(np.abs(line) ** 2)) for line in sinogram.T])
    rotation = np.argmax(r)
    angle = 90 - rotation
    print(f'Rotation H: {angle} degrees')
    return angle
    
def calculate_rotate_v(img):
    r = np.array([np.sqrt(np.mean(np.abs(line) ** 2)) for line in sinogram.T])
    rotation = np.argmax(r)
    angle = rotation
    print(f'Rotation V: {angle} degrees')
    return angle
```

```python
def rotate(image, angle, center = None, scale = 1.0):
    (h, w) = image.shape[:2]

    if center is None:
        center = (w / 2, h / 2)

    # Perform the rotation
    M = cv2.getRotationMatrix2D(center, angle, scale)
    rotated = cv2.warpAffine(image, M, (w, h))

    return rotated
```

```python
def view(img):
    plt.figure(figsize=(20, 20))
    plt.imshow(img, cmap='gray')
```

```python
def cv_rgb2gray(rgb):
    gray = np.mean(rgb, axis=2)
    return gray

def cv_binarization(gray_img, threshold = 100):
    binarized = 1.0 * (gray_img > threshold)
    return binarized

def cv_opening(img, kernel, iter_erosion= 1, iter_dilatation = 1):
    img = cv2.erode(img, kernel, iter_erosion)
    img = cv2.dilate(img, kernel, iter_dilatation)
    return img

def cv_closing(img, kernel, iter_erosion= 1, iter_dilatation = 1):
    img = cv2.dilate(img, kernel, iter_dilatation)
    img = cv2.erode(img, kernel, iter_erosion)
    return img

def kernel(n):
    kernel  = np.ones((n, n),np.uint8)
    return kernel

def kernel2(n, m):
    kernel  = np.ones((n, m),np.uint8)
    return kernel
```

```python
img  = cv2.imread('./image/table.jpg')
gray = cv_rgb2gray(img)
h, w = gray.shape
```

```python
sinogram = discrete_radon_transform(gray)
plt.imshow(sinogram)
```

```python
v = calculate_rotate_v(gray)
```

```python
view(gray) #   
```

```python
rh1 = rotate(gray, v  + 180)
view(rh1)
```

```python
hl = cv_closing(rh1, kernel2(1, 101))
view(hl)
```

```python
vl = cv_closing(rh1, kernel2(53, 1))
view(vl)
```
